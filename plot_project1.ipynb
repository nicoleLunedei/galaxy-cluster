# NON ISOTHERMAL CASE

```python
r, T_r, T_ICM = np.loadtxt("/home/claudia_pecorari/fortran/primo_progetto/temperature_gradT.dat", unpack=True)
r, p_num_gradT, p_an_gradT = np.loadtxt("/home/claudia_pecorari/fortran/primo_progetto/density_gradT.dat", unpack=True)
T_rebusco=7*((1+(r/71)**3)/(2.3+(r/71)**3))*1.16e7
```

Temperature profile: constant, normal and Rebusco

```python
plt.plot(r,T_r, label=r'$T(r)$')
plt.plot(r,T_ICM, label=r'$T_{ICM}$', ls='--', color='black')
plt.plot(r,T_rebusco, label=r'$T_{Rebusco}$', color='red')
plt.ylabel(r'$\log(T)(K)$')
plt.xlabel(r'$\log(r/kpc)$')
plt.xscale('log')
plt.yscale('log')
plt.title('Temperature profile')
plt.legend()
```

Gas density profile with BCG and temperature gradient:

```python
plt.plot(r, p_num_noBCG, label='numerical no BCG+isothermal '+ r'$(\rho_0=1.275*10^{-26})$', c='red', ls=':')
plt.plot(r, p_num_gradT, label='numerical BCG+dT/dr '+ r'$(\rho_0=6*10^{-26})$', c='black')
plt.plot(r, p_an_noBCG, label='NFW DM density', c='blue', ls='--')
plt.ylabel(r'$\log(\rho)$'+ ' (cgs?)', fontsize=15)
plt.xlabel(r'$\log(r)$'+ ' (kpc)', fontsize=15)
plt.xscale('log')
plt.yscale('log')
plt.title('Gad density profile with BCG and T gradient')
plt.legend(fontsize=8)
```
